```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsizes = 5
nlevels = 6
ranges = np.zeros((nlevels, nsizes))
```

```python
ranges[0][0] = 10

ranges[1][0] = 7
ranges[1][1] = 2
ranges[1][2] = 1

ranges[2][0] = 4
ranges[2][1] = 3
ranges[2][2] = 2
ranges[2][3] = 1

ranges[3][0] = 2
ranges[3][1] = 2
ranges[3][2] = 4
ranges[3][3] = 1
ranges[3][4] = 1

ranges[4][0] = 0
ranges[4][1] = 3
ranges[4][2] = 3
ranges[4][3] = 2
ranges[4][4] = 2

ranges[5][0] = 0
ranges[5][1] = 0
ranges[5][2] = 5
ranges[5][3] = 3
ranges[5][4] = 2
ranges = ranges.astype(int)
```

```python
for i in range(nlevels):
    for j in range(nsizes):
        if ranges[i, j] != 0:
            x = np.random.uniform(low=1, high=2, size=(ranges[i,j]))
            y = np.random.uniform(low=3*i+1, high=3*i+3, size=(ranges[i,j]))
            plt.scatter(x, y, color='c', s=40/(j+1), edgecolors='b')
plt.xlim(-4,8)
plt.savefig("res.pdf", format='pdf')
```
