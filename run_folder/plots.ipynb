```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma
import matplotlib as mpl
from matplotlib.colors import LogNorm
```

```python
res = np.loadtxt("data.txt")

print(res.shape)

time = 9

res = res.reshape((4, 100, 100))
```

```python
res[res<0.001] = 0.001
```

```python
f, ax = plt.subplots(2, 3, sharey='row', figsize=(15, 10))

cmap = plt.cm.jet
cmaplist = [cmap(i) for i in range(cmap.N)]
cmap = mpl.colors.LinearSegmentedColormap.from_list('Custom cmap', cmaplist, cmap.N)

im = ax[1,0].imshow(res[1],aspect='auto', cmap=cmap, norm=LogNorm(vmin=0.001, vmax=1))
ax[1,0].set_xticks(np.linspace(0,100, 6), np.linspace(0,5, 6), fontsize=20)
ax[1,0].set_xlabel("v", fontsize=25)
ax[1,0].set_yticks(np.linspace(0,99, 6), np.linspace(0,5, 6), fontsize=20)
ax[1,0].set_ylabel("x", fontsize=25)

ax[1,1].imshow(res[2],aspect='auto', cmap=cmap, norm=LogNorm(vmin=0.001, vmax=1))
ax[1,1].set_xticks(np.linspace(0,100, 6), np.linspace(0,5, 6), fontsize=20)
ax[1,1].set_xlabel("v", fontsize=25)
ax[1,2].imshow(res[3],aspect='auto', cmap=cmap, norm=LogNorm(vmin=0.001, vmax=1))
ax[1,2].set_xticks(np.linspace(0,100, 6), np.linspace(0,5, 6), fontsize=20)
ax[1,2].set_xlabel("v", fontsize=25)


ax[0,0].set_title("t=2.0", fontsize=25)
ax[0,0].plot(res[1][:-5, 1], label="v=0.2", lw=3)
ax[0,0].plot(res[1][:-5, 10], label="v=2.5", lw=3)
ax[0,0].plot(res[1][:-5, 40], label="v=3.5", lw=3)
ax[0,0].set_yticks(np.linspace(0,1, 6), np.linspace(0,1, 6).round(1), fontsize=20)
ax[0,0].set_ylabel("f(x,v,t)", fontsize=25)
ax[0,0].set_xticks(np.linspace(0,99, 6), np.linspace(0,5, 6), fontsize=20)
ax[0,0].set_xlabel("x", fontsize=25)
ax[0,0].grid()
ax[0,0].legend(fontsize=25)

ax[0,1].set_title("t=4.0", fontsize=25)
ax[0,1].plot(res[2][:-5, 1], label="v=0.2", lw=3)
ax[0,1].plot(res[2][:-5, 10], label="v=2.5", lw=3)
ax[0,1].plot(res[2][:-5, 40], label="v=3.5", lw=3)
ax[0,1].set_xticks(np.linspace(0,99, 6), np.linspace(0,5, 6), fontsize=20)
ax[0,1].set_xlabel("x", fontsize=25)
ax[0,1].grid()
ax[0,1].legend(fontsize=25)

ax[0,2].set_title("t=8.0", fontsize=25)
ax[0,2].plot(res[3][:-5, 1], label="v=0.2", lw=3)
ax[0,2].plot(res[3][:-5, 10], label="v=2.5", lw=3)
ax[0,2].plot(res[3][:-5, 40], label="v=3.5", lw=3)
ax[0,2].set_xticks(np.linspace(0,99, 6), np.linspace(0,5, 6), fontsize=20)
ax[0,2].set_xlabel("x", fontsize=25)
ax[0,2].grid()
ax[0,2].legend(fontsize=25)

cbar_ax = f.add_axes([0.095, 0.0, 0.82, 0.04])

cbar = f.colorbar(im, cax=cbar_ax, orientation="horizontal")
cbar.ax.tick_params(labelsize=25)
cbar.set_label("f(x,v,t)", fontsize=25)

plt.savefig("ball_const.pdf", format='pdf')
```
